Question 10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
with engine.connect() as connection:
    rs = connection.execute(
                    """SELECT * 
                        FROM (
                        SELECT name, SUM(CASE WHEN b.memid = 0 THEN b.slots * f.guestcost ELSE b.slots * f.membercost END) AS revenue
                        FROM Facilities AS f
                        INNER JOIN Bookings AS b
                        ON f.facid = b.facid
                        GROUP BY f.name
                        ) AS r
                        WHERE r.revenue < 1000
                        ORDER BY r.revenue;"""
                        )
    df = pd.DataFrame(rs.fetchall())
    df.columns = rs.keys()
```

 Question 11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
engine2 = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
with engine2.connect() as connection:
    rs2 = connection.execute("""
                    SELECT
                        m.firstname AS member_firstname, m.surname AS member_surname,
                        m1.firstname AS m_rec_firstname, m1.surname AS member_rec_firstname
                    FROM Members AS m
                    LEFT JOIN Members AS m1
                    ON m.recommendedby = m1.memid
                    ORDER BY member_surname, member_firstname
                    ;"""
                        )
    df2 = pd.DataFrame(rs2.fetchall())
    df2.columns = rs2.keys()
```

Question 12: Find the facilities with their usage by member, but not guests

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
with engine.connect() as connection:
    rs3 = connection.execute("""
                    SELECT f.name, COUNT(a.memid) AS member_usage, a.facid
                    FROM (
                        SELECT facid, memid
                        FROM Bookings
                        WHERE memid != 0 ) AS a
                    LEFT JOIN Facilities as f
                    ON a.facid = f.facid
                    GROUP BY a.facid
                    ;"""
                        )
    df3 = pd.DataFrame(rs3.fetchall())
    df3.columns = rs3.keys()
```
